<h1> ðŸ‘©â€ðŸ’» AlteraÃ§Ã£o dos Dados </h1>
<p>Ao verificarmos superficialmente os dados iremos realizar algumas alteraÃ§Ãµes destes.</p>

```python
def data_alteration(data):
    data = get_data(url)
    # Altear a coluna date para datetime
    data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
    
    # Classiticando a coluna yr_built que contÃ©m 0 para a data de 1970 
    data['yr_built'] = data['yr_built'].apply(lambda x: pd.to_datetime(x, format = '%Y') if x != 0 
                               else pd.to_datetime('1970-01-01'))
    
    # Construindo a coluna contendo os anos de edifÃ­cio
    data['yr_life'] = int(datetime.now().strftime('%Y')) - data['yr_built'].dt.year
    
    # Classificando o preÃ§o em levels
    data['level'] = data.price.apply(lambda x: 'level_0' if x <= data.price.quantile(0.25) else
                           'level_1' if (x > data.price.quantile(0.25)) & (x <= data.price.quantile(0.50)) else
                           'level_2' if (x > data.price.quantile(0.50)) & (x <= data.price.quantile(0.75)) else
                           'level_4')
    
    # Separando os meses em temporadas
    data['month'] = pd.to_datetime(data.date).dt.month
    data['season'] = data['month'].apply(lambda x: 'winter' if x <= 2 else
                                                    'spring' if (x > 2) & (x <= 5) else
                                                    'summer' if (x > 6) & (x <= 8) else
                                                    'fall')
    
    # Removendo dados duplicados e algumas colunas
    data.rename(columns = {'price_x': 'price', 'price_y': 'price_mean'}, inplace = True)
    data.drop(['sqft_living15', 'sqft_lot15'], axis = 1)
    
    return data
```

```python
new_data = data_alteration(data)
```

Para responder essa pergunta, iremos realizar as seguintes etapas:
<ul>
<li>Agrupar os dados por cep</li>
<li>Calcular a mÃ©dia dos preÃ§os por cep.</li>

Iremos explorar os outros cenÃ¡rios no intuito de verificar quais outros atributos que podem impactar em nossa anÃ¡lise. 

___
**Primeiro cenÃ¡rio:** nesta iremos realizar uma anÃ¡lise contendo a **condiÃ§Ã£o** do imÃ³vel e o **preÃ§o mÃ©dio** e verificar quantos imÃ³veis para compra restaram desta seleÃ§Ã£o.

```python
data_select = new_data[['zipcode', 'price']].groupby('zipcode').mean().reset_index()
df_select = pd.merge(new_data, data_select, how='inner', on = 'zipcode')
df_select.rename(columns = {'price_x': 'price', 'price_y': 'price_mean'}, inplace = True)
```

```python
for i in range(len(df_select)):
    if (df_select.loc[i, 'price'] < df_select.loc[i, 'price_mean']) & (df_select.loc[i, 'condition'] >= 3):
        df_select.loc[i, 'Proposition'] = 'Buy'
    else:
        df_select.loc[i, 'Proposition'] = 'Not Buy'
```

<h2>Quais imÃ³ves que deverÃ­amos vender e por qual preÃ§o?</h2>

Para respondermos essa pergunta iremos primeiro verificar o sobre os preÃ§os dos imÃ³veis

```python
df_profit = df_select[df_select['Proposition'] == 'Buy']
```

```python
plt.figure(figsize = (16,7))
plt.xticks(rotation = 60)
sns.barplot(data = df_profit, x = df_profit['zipcode'], y = df_profit['price']);
```

Vemos com isso que alguns imÃ³veis contÃ©m um preÃ§o bem elevado e outros nÃ£o tanto assim.

Outro aspecto que iremos verificar Ã© se a distribuiÃ§Ã£o de preÃ§o se mantÃ©m a mesma ou se ela sofre alteraÃ§Ãµes conforme o tempo, com isso realizaremos o seguinte grÃ¡fico abaixo: 

```python
plt.figure(figsize = (14,7))
sns.lineplot(data = df_profit, x = 'season', y = 'price');
```

Vimos que durante o tempo os preÃ§os dos imÃ³veis sofrem flutuaÃ§Ãµes significativas, em relaÃ§Ã£o a isso, iremos tomar duas decisÃµes:
<ul>
<li>Caso os <strong>preÃ§os</strong> sejam menor que a <strong>mediana</strong>, iremos <strong>acrescentar</strong> um lucro em <strong>30%</strong> em relaÃ§Ã£o ao preÃ§o de compra.</li>
PreÃ§o venda < Mediana = PreÃ§o * 1,30 <br><br>
<li>Caso os <strong>preÃ§os</strong> sejam maiior que a <strong>mediana</strong>, iremos <strong>acrescentar</strong> um lucro de <strong>10%</strong> em relaÃ§Ã£o ao preÃ§o de compra.</li>
PreÃ§o venda < Mediana = PreÃ§o * 1,10 <br><br>
</ul>

```python
# Agrupando os dados de acordo com o cep, Ã©poca e o preÃ§o
df2 = df_profit[['zipcode','season','price']].groupby(['zipcode','season']).median().reset_index()
df2 = df2.rename(columns={'price': 'price_median_season'})

# unir df2 com df_profit
df_final = pd.merge(df_profit, df2, how ='inner', on=['zipcode','season'])
df_final
```

```python
for i in range(len(df_final)):
    if df_final.loc[i, 'price'] < df_final.loc[i, 'price_median_season']:
        df_final.loc[i, 'sale_price'] = df_final.loc[i, 'price'] * 1.3
    else:
        df_final.loc[i, 'sale_price'] = df_final.loc[i, 'price'] * 1.1
```

```python
df_final['profit'] = df_final['sale_price'] - df_final['price']
```

```python
df_final
```

O lucro final junto com os custos finais podem ser resumidos com a seguinte tabela:

```python
profit = df_final.profit.sum()
sell = df_final.sale_price.sum()
cost = df_final.price.sum()

cost_profit = pd.DataFrame({'Sell': [sell], 'Cost': [cost], 'Profit': [profit]})
cost_profit
```

Nisso vemos que o temos um preÃ§o de vendas de ***6.322.951.472.30***, um custo de ***5.345.126.211.00*** gerando um lucro final de ***977.825.261.30***
